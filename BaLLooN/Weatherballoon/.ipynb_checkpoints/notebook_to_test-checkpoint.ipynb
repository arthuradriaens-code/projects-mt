```python
import radiopropa
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,-100)
X
```

```python
N = np.sin(X/50)
N
```

```python
ice_model = radiopropa.IceModel_Data1D()
ice_model.setDataFromVector(radiopropa.DoubleVector_1D(X),radiopropa.DoubleVector_1D(N))
```

```python
N_ = []
for x in X:
    N_.append(ice_model.getValue(radiopropa.Vector3d(0,0,x)))
```

```python
plt.scatter(X,N,c='blue')
plt.scatter(X,N_,c='orange')
plt.show()
```

```python
plt.scatter(X,N_-N,c='black')
plt.show()
```

```python
step = X[1]-X[0]
Z = X + step/2
Z
```

```python
N__ = []
for z in Z:
    N__.append(ice_model.getValue(radiopropa.Vector3d(0,0,z)))
```

```python
plt.scatter(X,N_,c='orange')
plt.scatter(Z,N__,c='red')
plt.show()
```
